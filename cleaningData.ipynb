```python
import pandas as pd

df = pd.read_csv("used_motorcycle.csv")
```

```python
df.head()
```

```python
df = df[["CarName", "YEAR", "COLOR", "mileage", "price"]]
df = df.rename({"CarName": "Motorcycle"}, axis=1)
df = df.rename({"COLOR": "Color"}, axis=1)
df = df.rename({"YEAR": "Year"}, axis=1)
df = df.rename({"price": "Price"}, axis=1)
df = df.rename({"mileage": "Mileage"}, axis=1)
df.head()
```

```python
df = df[df["Price"].notnull()]
df.head()
```

```python
df.isnull().sum()
```

```python
average_mileage = df['Mileage'].mean()
df['Mileage'].fillna(average_mileage, inplace=True)
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
df.isnull().sum()
```

```python
color_mapping = {
    'สีขาว': 'white',
    'สีดำ': 'black',
    'สีแดง': 'red',
    'สีเขียว': 'green',
    'สีน้ำตาล': 'brown',
    'สีชมพู': 'pink',
    'สีน้ำเงิน': 'blue',
    'สีเทา': 'gray',
    'สีส้ม': 'orange',
    'สีเหลือง': 'yellow',
    'สีม่วง': 'purple',
    'สีฟ้า': 'blue',
}

# Function to clean and translate the colors
def clean_and_translate_color(thai_color):
    for thai, eng in color_mapping.items():
        if thai in thai_color:
            return eng + thai_color.replace(thai, '')
    return thai_color

df['Color_new'] = df['Color'].apply(clean_and_translate_color)
```

```python
df = df[df['Price'] != 0.0]
```

```python
def clean_color_remove_thai(x):
    if 'white' in x:
        return 'white'
    if 'black' in x:
        return 'black'
    if 'red' in x:
        return 'red'
    if 'green' in x:
        return 'green'
    if 'brown' in x:
        return 'brown'
    if 'pink' in x:
        return 'pink'
    if 'blue' in x:
        return 'blue'
    if 'gray' in x:
        return 'gray'
    if 'orange' in x:
        return 'orange'
    if 'yellow' in x:
        return 'yellow'
    if 'purple' in x:
        return 'purple'
    if 'blue' in x:
        return 'blue'
    return 'Other'

df['Color_new'] = df['Color_new'].apply(clean_color_remove_thai)
```

```python
df = df[["Motorcycle", "Color_new", "Year", "Mileage", "Price"]]
df = df.rename({"Color_new": "Color"}, axis=1)
```

```python
df['Motorcycle'] = df['Motorcycle'].str.upper()
```

```python
df['MotorcycleStrip'] = df['Motorcycle'].str.strip()
df
```

```python
# Count the occurrences of each value in the 'Motorcycle' column
value_counts = df['MotorcycleStrip'].value_counts()

# Filter the DataFrame to remove rows with counts less than 5
df = df[df['MotorcycleStrip'].map(value_counts) >= 5]
```

```python
def clean_name(x):
    if 'WAVE 125 I' in x or 'WAVE125 I' in x:
        return 'WAVE125I'
    if 'PCX 150' in x or 'NEW PCX150' in x:
        return 'PCX150'
    if 'SCOOPY I' in x or 'SCOOPY-I S12' in x:
        return 'SCOOPY-I'
    if 'CLICK 125I' in x:
        return 'CLICK125I'
    if 'CLICK 110I' in x:
        return 'CLICK110I'
    if 'WAVE 110I' in x or 'WAVE110 I' in x or 'WAVE 110 I' in x:
        return 'WAVE110I'
    if 'GRAND FILANO HYBRID' in x or 'GRAND FILANO HYBRID ABS' in x:
        return 'GRAND FILANO'
    return x

# df['Motorcycle'] = df['MotorcycleStrip'].apply(clean_name)
df_copy = df.copy()

# Apply your function to the 'MotorcycleStrip' column of the copy
df_copy['Motorcycle'] = df_copy['MotorcycleStrip'].apply(clean_name)
```
